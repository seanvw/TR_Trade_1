Write a python application to make calculations of profit as APR for each Share held in a Portfolio.

Include a demonstation of it's usage.

You can use the current share price for a given ticker symbol using yfinance.

A share can be bought and sold within a day or over a longer period, months or years.

When a share is bought, the date and time is captured to the minute.

When a share is sold, the date and time is captured to the minute.

The application should be able to calculate APR based on a combination of sold shares (realised profit/loss) together with the current value of shares currently held within a portfolio

The application should report the realised profit/loss of shares no longer owned, together with its bought and sold datetimes and the days elapsed between purchase and sale. Additionally the effective APR of the share sale should be reported.

The application should report the unrealised profit/loss of shares still held, together with its bought date and time. Additionally the effective APR of the share sale should be reported.

Within the application, all date and times should be human readable

The demonstation of usage should give examples of realised share dealings held for 1 day, 5 days and 6 months respectively

The demonstation of usage should give examples of unrealised share purchases

Respond with the code only! Nothing else!

```python

import yfinance as yf
from datetime import datetime, timedelta

class ShareTransaction:
    def __init__(self, ticker, shares, buy_price, buy_datetime, sell_price=None, sell_datetime=None):
        self.ticker = ticker
        self.shares = shares
        self.buy_price = buy_price
        self.buy_datetime = buy_datetime
        self.sell_price = sell_price
        self.sell_datetime = sell_datetime
        self.realised = sell_price is not None

    def calculate_profit_loss(self):
        if self.realised:
            return (self.sell_price - self.buy_price) * self.shares
        else:
            current_price = yf.Ticker(self.ticker).info['currentPrice']
            return (current_price - self.buy_price) * self.shares

    def calculate_duration(self):
        if self.realised:
            return (self.sell_datetime - self.buy_datetime).days
        else:
            return (datetime.now() - self.buy_datetime).days

    def calculate_apr(self):
        if self.realised:
            days = (self.sell_datetime - self.buy_datetime).days
        else:
            days = (datetime.now() - self.buy_datetime).days

        if days == 0:
            days = 1  # Avoid division by zero

        profit = self.calculate_profit_loss()
        cost = self.buy_price * self.shares
        apr = (profit / cost) * (365 / days) * 100
        return apr

class Portfolio:
    def __init__(self):
        self.transactions = {
            'realised': [],
            'unrealised': []
        }

    def add_transaction(self, transaction):
        if transaction.realised:
            self.transactions['realised'].append(transaction)
        else:
            self.transactions['unrealised'].append(transaction)

    def report(self):
        print("\nRealised Transactions:")
        print("------------------------")
        for i, txn in enumerate(self.transactions['realised'], 1):
            print(f"Transaction {i}:")
            print(f"Ticker: {txn.ticker}")
            print(f"Shares: {txn.shares}")
            print(f"Buy Price: ${txn.buy_price:.2f}")
            print(f"Buy DateTime: {txn.buy_datetime.strftime('%Y-%m-%d %H:%M:%S')}")
            print(f"Sell Price: ${txn.sell_price:.2f}")
            print(f"Sell DateTime: {txn.sell_datetime.strftime('%Y-%m-%d %H:%M:%S')}")
            print(f"Days Held: {txn.calculate_duration()}")
            print(f"Profit/Loss: ${txn.calculate_profit_loss():.2f}")
            print(f"APR: {txn.calculate_apr():.2f}%")
            print("------------------------")

        print("\nUnrealised Transactions:")
        print("-------------------------")
        for i, txn in enumerate(self.transactions['unrealised'], 1):
            print(f"Transaction {i}:")
            print(f"Ticker: {txn.ticker}")
            print(f"Shares: {txn.shares}")
            print(f"Buy Price: ${txn.buy_price:.2f}")
            print(f"Buy DateTime: {txn.buy_datetime.strftime('%Y-%m-%d %H:%M:%S')}")
            print(f"Current Price: ${yf.Ticker(txn.ticker).info['currentPrice']:.2f}")
            print(f"Days Held: {txn.calculate_duration()}")
            print(f"Profit/Loss: ${txn.calculate_profit_loss():.2f}")
            print(f"APR: {txn.calculate_apr():.2f}%")
            print("-------------------------")

def main():
    portfolio = Portfolio()

    # Example 1: 1 day holding period
    txn1 = ShareTransaction(
        'AAPL',
        10,
        150.0,
        datetime.now() - timedelta(days=1),
        155.0,
        datetime.now()
    )
    portfolio.add_transaction(txn1)

    # Example 2: 5 days holding period
    txn2 = ShareTransaction(
        'MSFT',
        5,
        200.0,
        datetime.now() - timedelta(days=5),
        210.0,
        datetime.now()
    )
    portfolio.add_transaction(txn2)

    # Example 3: 6 months holding period
    txn3 = ShareTransaction(
        'GOOG',
        2,
        2500.0,
        datetime.now() - timedelta(days=180),
        2700.0,
        datetime.now()
    )
    portfolio.add_transaction(txn3)

    # Example 4: Unrealised transaction
    txn4 = ShareTransaction(
        'AMZN',
        1,
        3000.0,
        datetime.now() - timedelta(days=30)
    )
    portfolio.add_transaction(txn4)

    portfolio.report()

if __name__ == "__main__":
    main()
```
