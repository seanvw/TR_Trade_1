With Python object orientation write classes that support the following requirements in the domain of Share dealing of a Portfolio. Also include some demonstration of usage with some dummy data.

A share can be bought and sold within a day or over a longer period, months or years.

When a share is bought, the date and time is captured to the minute.

When a share is sold, the date and time is captured to the minute.

The application should be able to calculate APR based on a combination of sold shares (realised profit/loss) together with the current value of shares currently held within a portfolio

The application should report the realised profit/loss of shares no longer owned, together with its bought and sold datetimes and the days elapsed between purchase and sale. Additionally the effective APR of the share sale should be reported.

The application should report the unrealised profit/loss of shares still held, together with its bought date and time. Additionally the effective APR of the share sale should be reported.

Within the application, all date and times should be human readable

The demonstation of usage should give examples of realised share dealings held for 1 day, 5 days and 6 months respectively

The demonstation of usage should give two examples of unrealised share purchases, one is a profit and one is a loss

The details of realised and unrealised dealing should be pretty printed

Respond with the code only! Nothing else!

```python

import datetime

class Share:
    def __init__(self, symbol, purchase_price, quantity):
        self.symbol = symbol
        self.purchase_price = purchase_price
        self.quantity = quantity
        self.purchase_datetime = datetime.datetime.now()
        self.sale_datetime = None
        self.sale_price = None

    def sell(self, sale_price, quantity):
        if quantity > self.quantity:
            raise ValueError("Cannot sell more shares than owned.")
        self.sale_datetime = datetime.datetime.now()
        self.sale_price = sale_price
        self.quantity -= quantity

    def get_profit_loss(self):
      if self.sale_datetime is None:
        return None
      return (self.sale_price - self.purchase_price) * self.quantity

    def get_apr(self):
      if self.sale_datetime is None:
        return None
      time_delta = self.sale_datetime - self.purchase_datetime
      days = time_delta.days
      if days == 0:
          days = 1
      profit_loss = self.get_profit_loss()
      return (profit_loss / (self.purchase_price * self.quantity)) * (365/days) * 100

    def __str__(self):
        return f"Symbol: {self.symbol}, Quantity: {self.quantity}, Purchase Price: {self.purchase_price}, Purchase Datetime: {self.purchase_datetime.strftime('%Y-%m-%d %H:%M')}"


class Portfolio:
    def __init__(self):
        self.shares = []

    def buy_share(self, share):
        self.shares.append(share)

    def sell_share(self, symbol, sale_price, quantity):
        for share in self.shares:
            if share.symbol == symbol:
                share.sell(sale_price, quantity)
                return
        raise ValueError(f"Share with symbol {symbol} not found in portfolio.")

    def get_portfolio_value(self, current_prices):
      total_value = 0
      for share in self.shares:
        if share.sale_datetime is None:
          total_value += current_prices.get(share.symbol, 0) * share.quantity
      return total_value


    def report(self, current_prices):
        print("Realised Profit/Loss:")
        for share in self.shares:
          if share.sale_datetime:
              profit_loss = share.get_profit_loss()
              apr = share.get_apr()
              time_delta = share.sale_datetime - share.purchase_datetime
              print(f"  Symbol: {share.symbol}, Profit/Loss: {profit_loss:.2f}, APR: {apr:.2f}%, Bought: {share.purchase_datetime.strftime('%Y-%m-%d %H:%M')}, Sold: {share.sale_datetime.strftime('%Y-%m-%d %H:%M')}, Days Held: {time_delta.days}")
        print("\nUnrealised Profit/Loss:")
        for share in self.shares:
            if share.sale_datetime is None:
                current_price = current_prices.get(share.symbol, share.purchase_price)
                profit_loss = (current_price - share.purchase_price) * share.quantity
                time_delta = datetime.datetime.now() - share.purchase_datetime
                days = time_delta.days
                if days == 0:
                    days = 1
                apr = (profit_loss / (share.purchase_price * share.quantity)) * (365/days) * 100
                print(f"  Symbol: {share.symbol}, Profit/Loss: {profit_loss:.2f}, APR: {apr:.2f}%, Bought: {share.purchase_datetime.strftime('%Y-%m-%d %H:%M')}")


# Demonstration
portfolio = Portfolio()


portfolio.buy_share(Share("MSFT", 150, 10))
portfolio.buy_share(Share("GOOG", 2500, 2))
portfolio.buy_share(Share("AAPL", 170, 5))

# Simulate selling shares
portfolio.sell_share("MSFT", 160, 10)

#Simulate share held for 5 days
import time
time.sleep(5)
portfolio.buy_share(Share("TSLA", 800, 1))
portfolio.sell_share("TSLA", 810, 1)

#Simulate share held for 6 months
import datetime
six_months_ago = datetime.datetime.now() - datetime.timedelta(days=180)

class Share:
    def __init__(self, symbol, purchase_price, quantity, purchase_datetime=datetime.datetime.now()):
        self.symbol = symbol
        self.purchase_price = purchase_price
        self.quantity = quantity
        self.purchase_datetime = purchase_datetime
        self.sale_datetime = None
        self.sale_price = None

    def sell(self, sale_price, quantity):
        if quantity > self.quantity:
            raise ValueError("Cannot sell more shares than owned.")
        self.sale_datetime = datetime.datetime.now()
        self.sale_price = sale_price
        self.quantity -= quantity


portfolio.buy_share(Share("NVDA", 200, 5, purchase_datetime=six_months_ago))
portfolio.sell_share("NVDA", 260, 5)

current_prices = {"GOOG": 2600, "AAPL": 180}
portfolio.report(current_prices)
```
